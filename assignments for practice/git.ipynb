## Задание 3

откатывание к HEAD. Откатывание к любому другому коммиту. Удаление коммитов

Чтобы перейти к определенному коммиту, используйте команду `git checkout <SHA коммита>`. Основная функция `git checkout` - перемещение указателя HEAD в вашей локальной копии проекта. Указатель HEAD может быть перемещён на любой коммит в истории, либо на любой дрйгой комит из любой ветки.

В результате перехода к конкретному коммиту вы оказываетесь в «состоянии отсоединенного указателя HEAD» («detached HEAD state»). В общем, HEAD отклонился от известных веток, так что все изменения с этого момента будут формировать новую последовательность изменений в истории git. При этом git выведет вам напутственное сообщение:

«Вы в состоянии detached HEAD. Можете осмотреться, внести какие-нибудь экспериментальные изменения и закоммитить их. Если затем вы переключитесь обратно на ветку, вы сбросите все коммиты, сделанные в этом состоянии, никак не задев никакие ветки».

Исходя из этого, у вас есть два варианта:

   1. поэкспериментировать, а затем сбросить все изменения, вернувшись в предыдущую ветку
   2. продолжить работу с этого места и начать новую ветку из этой точки.
   
Для того, чтобы переместить указатель к последнему коммиту в origin воспользуйтесь командой `git checkout <имя ветки>`, либоб если вы забыли имя ветки и не можете вспомнить, как её посмотреть (`git branch`), можно использовать команду `git checkout -`.

Команда `git checkout HEAD -- <путь до файла> ` позволяет отменить индексированные изменения для конкретного файла до последнего коммита (на который указывает HEAD). `--` означает "рассматривать каждый аргумент после -- как имя файла".

В процессе работы вы можете столкнуться с тем, что в результате ваших экспериментов с git'ом все изменения куда-то пропали. Для решения этой проблемы бывает полезна команда `git reflog`, которая выводит все промежуточные состояния файлов. Для того, чтобы переключиться на интересующее вас состояние воспользуйтесь командой `git reset --hard HEAD@{<порядковый номер из git reflog>}`, либо `git reset --hard <хэш-идентификатор промежуточного состояния>`. Команда `git reset --hard <SHA коммита>`, может быть применена для перемещения указателя HEAD *и указателя Main* к необходимому коммиту, затем происходит сброс раздела проиндексированных файлов и рабочего каталога до указанного коммита. Все предыдущие ожидающие изменения в разделе проиндексированных файлов и рабочем каталоге сбрасываются в соответствии с состоянием дерева коммитов. Это значит, что любая работа, находившаяся в состоянии ожидания в разделе проиндексированных файлов и рабочем каталоге, будет потеряна. Таким образом, если команда `git checkout HEAD -- <путь до файла>` позволяет вам удалить данные об изменениях файлов (перевести файлы из staged в unmodified) и вернуть их в изначальное состояние, то команда `git reset --hard` позволяет вам удалять добавленные коммиты.

На первый взгляд `git reset` и `git checkout` очень похожи, однако есть между ними существенное различие:

- `git reset` перемещает указатель HEAD и указатель текущей ветки.

- `git checkout` работает исключительно с указателем HEAD; В результате переещения указателя вы получите detached HEAD.

<center><img src="imgs/reset_and_checkout.png"></center>

Ещё один полезный приём при работе с git: если в вашей ветке уже было внесено много изменений, а вам необходимо переключиться на другой коммит или другую ветку, вы можете использовать команду `git stash save`, чтобы быстро сохранить все изменения, а затем при помощи команды `git stash pop` восстановить эти изменения.

При помощи `git checkout` переключитесь на один из ранних коммитов. Изменились ли файлы в директории с проектом? Как выглядит `git log` после перемещения указателя? Вернитесь на последний коммит в вашей ветке.


## Задание 4

В определённый момент работы над проектом вы захотите синхронизировать ваши локальные изменения с удалённым репозиторием проекта на сервере. Таких репозиториев может быть несколько: например, у вас есть публичный репозиторий, куда вы выкладываете стабильные версии вашей програмы и отдельный приватный репозиторий для разработки.

Для просмотра достпуных удалённых подключениях к репозиториям воспользуйтесь командой `git remote -v`

В настоящий момент вам доступен только один удалённый репозиторий, указывающий на созданный вами проект (*origin*). В процессе клонирования с помощью команды `git clone` автоматически создается удаленное подключение к исходному репозиторию (такое соединение и называется *origin*).

Для того чтобы по необзодимости получать обновления основного проекта курса, давайте добавим к списку удалённых подключений соответствующий удалённый репозиторий (https://gitlab.sirius-web.org/courses/python-for-beginners-2022). Для этого воспользуемся командой `git remote add upstream https://gitlab.sirius-web.org/courses/python-for-beginners-2022.git`. В данной команде указывается название подключения *upstream* (ярлык, подменяющий собой длинный url-адрес) и *url*. Проверьте список достпуных подключений.

<a href="https://www.atlassian.com/ru/git/tutorials/syncing">Подробнее об использовании git remote</a>

Перед отправкой ваших изменений на сервер необходимо запросить текущее состояние проекта на сервере для этого воспользуйтесь командой `git fetch`. Для слияния ваших локальных изменений в проекте с удалённоё версией воспользуйтесь командой `git merge`. Зачастую при синхронизации изменений удобнее пользоваться командой `git pull`, которая заменяет собой последовательность команд `git fetch` и `git merge`.

По-умолчанию при использовании команды `git pull` данные об изменениях будут запрашиваться с удалённого репозитория помеченного как *origin* (см. `git remote -v`), однако можно явным образом указывать удалённый репозиторий, к которому мы хотим обратиться, например, `git pull <имя репозитория>`, например, `git fetch upstream`.

Синзронизируйте ваши локальные изменения проекта с удалённым репозиторием (origin). 

## Задание 5

Поскольку на одном проекте могут работать несколько разработчиков, ситуации, когда несколько человек модифицировали один и тот же файл повсеместны, что приводит к конфликтам при объединении (merge) изменений. Зачастую при синхронизации изменений git разрешает конфликты в автоматическом режиме и создаёт соответствующий коммит слияния. Однако вы можете столкнуться с cитуацией, когда git не создал коммит слияния автоматически. Он остановил процесс до тех пор, пока вы не разрешите конфликт. Чтобы в любой момент после появления конфликта увидеть, какие файлы не объединены, используйте команду `git status`

Всё, где есть неразрешённые конфликты слияния, перечисляется как неслитое. В конфликтующие файлы Git добавляет специальные маркеры конфликтов, чтобы вы могли исправить их вручную. Например, в результате слияния вашей ветки *iss53* c веткой *master* в вашем файле *index.html* появился раздел, выглядящий примерно так:

<<<<<<< HEAD:index.html<br>
Какой-то текст<br>
=======<br>
Какой-то текст<br>
Ещё немного какого-то текста<br>
\>>>>>>> iss53:index.html

Это означает, что версия из HEAD (вашей ветки *master*, то что вы извлекли из удалённого репозитория для слияния) — это верхняя часть блока (всё, что над `=======`), а версия из вашей ветки *iss53* представлена в нижней части. 
Чтобы разрешить конфликт, придётся выбрать один из вариантов, либо объединить содержимое по-своему. Например, вы можете разрешить конфликт, заменив весь блок следующим:

Какой-то текст<br>
Ещё немного какого-то текста<br>

В этом разрешении есть немного от каждой части, а строки `<<<<<<<`, `=======` и `>>>>>>>` полностью удалены. Разрешив каждый конфликт во всех файлах, запустите `git add` для каждого файла, чтобы отметить конфликт как решённый. Добавление файла в индекс означает для Git, что все конфликты в нём исправлены.

Попробуем разрешить конфликт. Для этого необходимо открыть текстовый файл `file_for_conflicts.txt` и удалить в нём лишнюю строчку (очевидно какую). Я, в свою очередь, внесу свои изменения в этот файл и синхронизирую их с `upstream`. Таким образом, при попытке выполнить `git pull upstream`, вы столкнётесь с конфликтом, который не может быть разрешён git в автоматическом режиме. При помощи команды `git status` проверьте статус файлов, явившихся причиной конфликта. Данный конфликт необходимо разрешить в ручном режиме. Откройте данный файл, разрешите конфликт, оставив только желаемую часть текста и удалив слежебные строки, и зафиксируйте слияние в коммите. Синхронизируйте внесённые вами изменения с вашим `origin`.
